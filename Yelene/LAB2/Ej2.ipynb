```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Cargar el dataset desde el archivo CSV
df = pd.read_csv('d4332a3056f44e1a1dec9600a31f21c8_boston.csv')

# Mostrar información básica del dataset
print(df.info())
print(df.describe())
print(df.isnull().sum())

# Identificar la columna problemática (por ejemplo, 'Location')
print(df['TOWN'].unique())

# Aplicar One-Hot Encoding a la columna 'TOWN'
df_encoded = pd.get_dummies(df, columns=['TOWN'], drop_first=True)

# Verificar los cambios
print(df_encoded.head())
print(df_encoded.columns)

# Imputar valores NaN con la media de la columna
df_imputed = df_encoded.fillna(df_encoded.mean())

# Verificar los cambios
print(df_imputed.info())
print(df_imputed.isnull().sum())

# Separar las características de la variable objetivo

X = df_imputed.drop('MEDV', axis=1)
y = df_imputed['MEDV']

# Escalar las características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Obtener los nombres de las columnas desde el DataFrame original
column_names = X.columns

# Crear DataFrames para los conjuntos de entrenamiento y prueba
train_data = pd.DataFrame(X_train, columns=column_names)
train_data['MEDV'] = y_train
test_data = pd.DataFrame(X_test, columns=column_names)
test_data['MEDV'] = y_test

# Mostrar las primeras filas de los conjuntos de datos de entrenamiento y prueba
print(train_data.head())
print(test_data.head())

# Guardar los conjuntos de datos procesados en archivos CSV
train_data.to_csv('boston_train_data.csv', index=False)
test_data.to_csv('boston_test_data.csv', index=False)
```
